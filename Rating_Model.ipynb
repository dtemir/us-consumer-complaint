# Define the rating model

```python
def rateme(df1):
  Ratings =300
  if df1['Consumer disputed?'] == 'Yes':
    if Rating>300:
      Ratings = Ratings - (0.2 * 300)
    else:
      Ratings = Ratings - (0.2 * Ratings)
  elif df1['Consumer disputed?'] == 'No':
    if Ratings>300:
      Ratings = Ratings + (0.15*Ratings)
    else:
      Ratings = Ratings + (0.15*300)
  else:
    Ratings = Ratings + (0.01*Ratings)


  if df1['Company response to consumer'] == 'Closed with non-monetary relief':
    if Rating>300:
      Ratings = Ratings - (0.01 * 300)
    else:
      Ratings = Ratings - (0.01 * Ratings)
  elif df1['Company response to consumer'] == 'Closed with explanation':
    if Rating>300:
      Ratings = Ratings - (0.01 * 300)
    else:
      Ratings = Ratings - (0.01 * Ratings)
  elif df1['Company response to consumer'] == 'Untimely response':
    if Rating>300:
      Ratings = Ratings - (0.03 * 300)
    else:
      Ratings = Ratings - (0.03 * Ratings)
  elif df1['Company response to consumer'] == 'In progress':
    if Rating>300:
      Ratings = Ratings - (0.01 * 300)
    else:
      Ratings = Ratings - (0.01 * Ratings)
  elif df1['Company response to consumer'] == 'Closed with monetary relief':
    if Ratings>300:
      Ratings = Ratings + (0.015*Ratings)
    else:
      Ratings = Ratings + (0.015*300)
  elif df1['Company response to consumer'] == 'Closed':
    if Rating>300:
      Ratings = Ratings - (0.01 * 300)
    else:
      Ratings = Ratings - (0.01 * Ratings)
  elif df1['Company response to consumer'] == 'Closed with relief':
    if Ratings>300:
      Ratings = Ratings + (0.015*Ratings)
    else:
      Ratings = Ratings + (0.015*300)
  elif df1['Company response to consumer'] == 'Closed without relief':
    if Rating>300:
      Ratings = Ratings - (0.01 * 300)
    else:
      Ratings = Ratings - (0.01 * Ratings)


  if df1['Timely Response'] == 'No':
    if Rating>300:
      Ratings = Ratings - (0.09 * 300)
    else:
      Ratings = Ratings - (0.09 * Ratings)
  elif df1['Timely Response'] == 'No':
    if Ratings>300:
      Ratings = Ratings + (0.065*Ratings)
    else:
      Ratings = Ratings + (0.065*300)
  return Ratings
```

```python
def rateme(df):
  Ratings =500
  if len(df)<50:
    return ("Too few Issues to have a score, keep it up")
  else:
    for key, df1 in df.iterrows():
      if df1['Consumer disputed?'] == 'Yes':
        if Ratings>300:
          Ratings = Ratings - (0.004 * 100)
        else:
          Ratings = Ratings - (0.004 * Ratings)
      elif df1['Consumer disputed?'] == 'No':
        if Ratings>300:
          Ratings = Ratings + (0.0094*Ratings)
        else:
          Ratings = Ratings + (0.0094*300)
      else:
        Ratings = Ratings + (0.009*Ratings)


      if df1['Company response to consumer'] == 'Closed with non-monetary relief':
        if Ratings>300:
          Ratings = Ratings - (0.01 * 100)
        else:
          Ratings = Ratings - (0.01 * Ratings)
      elif df1['Company response to consumer'] == 'Closed with explanation':
        if Ratings>300:
          Ratings = Ratings + (0.01 * 100)
        else:
          Ratings = Ratings + (0.01 * Ratings)
      elif df1['Company response to consumer'] == 'Untimely response':
        if Ratings>300:
          Ratings = Ratings - (0.02 * 100)
        else:
          Ratings = Ratings - (0.02 * Ratings)
      elif df1['Company response to consumer'] == 'In progress':
        if Ratings>300:
          Ratings = Ratings - (0.005 * 100)
        else:
          Ratings = Ratings - (0.005 * Ratings)
      elif df1['Company response to consumer'] == 'Closed with monetary relief':
        if Ratings>300:
          Ratings = Ratings + (0.015*Ratings)
        else:
          Ratings = Ratings + (0.015*300)
      elif df1['Company response to consumer'] == 'Closed':
        if Ratings>300:
          Ratings = Ratings - (0.0051 * 100)
        else:
          Ratings = Ratings - (0.0051 * Ratings)
      elif df1['Company response to consumer'] == 'Closed with relief':
        if Ratings>300:
          Ratings = Ratings + (0.0009*Ratings)
        else:
          Ratings = Ratings + (0.0009*300)
      elif df1['Company response to consumer'] == 'Closed without relief':
        if Ratings>300:
          Ratings = Ratings - (0.01 * 100)
        else:
          Ratings = Ratings - (0.01 * Ratings)


      if df1['Timely response?'] == 'No':
        if Ratings>300:
          Ratings = Ratings - (0.03 * 100)
        else:
          Ratings = Ratings - (0.03 * Ratings)
      elif df1['Timely response?'] == 'No':
        if Ratings>300:
          Ratings = Ratings + (0.001*Ratings)
        else:
          Ratings = Ratings + (0.001*300)
    return Ratings
```
